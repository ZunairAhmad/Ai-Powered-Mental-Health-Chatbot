```python
import torch
from tqdm.auto import tqdm

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
tokenizer.pad_token = tokenizer.eos_token

test_dff = test_dff.copy()
predicted_answers = []

for idx, row in tqdm(test_dff.iterrows(), total=len(test_dff), desc="Generating"):
    
    question = row["Context"]

    # correct prompt for base Llama-3.2
    manual_prompt = f"<s>Context: {question}\nResponse:"

    inputs = tokenizer(manual_prompt, return_tensors="pt").to(device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=150,
            do_sample=True,
            top_p=0.9,
            temperature=0.7,
            pad_token_id=tokenizer.eos_token_id
        )

    # FULL decode — do NOT slice token IDs for base model
    decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Correct extraction
    if "Response:" in decoded:
        answer = decoded.split("Response:")[1].strip()
    else:
        answer = decoded.strip()

    predicted_answers.append(answer)

test_dff["Predicted_Response"] = predicted_answers
```

```python
import evaluate

bleu = evaluate.load("bleu")

preds = test_dff["Predicted_Response"].astype(str).tolist()
refs  = test_dff["Response"].astype(str).tolist()

# BLEU needs references wrapped in a list
references = [[r] for r in refs]

bleu_score = bleu.compute(predictions=preds, references=references)

print("BLEU Score:", bleu_score["bleu"])
```

**Cosine Similarity**

```python
from sentence_transformers import SentenceTransformer, util
import torch

# Load embedding model
model_sts = SentenceTransformer('all-MiniLM-L6-v2')

preds = test_dff["Predicted_Response"].astype(str).tolist()
refs  = test_dff["Response"].astype(str).tolist()

# Compute embeddings
emb_preds = model_sts.encode(preds, convert_to_tensor=True)
emb_refs  = model_sts.encode(refs, convert_to_tensor=True)

# Pairwise cosine similarity for each row
cos_sim = util.cos_sim(emb_preds, emb_refs).diag().cpu().numpy()

# Save as column
test_dff["Cosine_Similarity"] = cos_sim

print("Cosine similarity column created!")
```

```python
import seaborn as sns

plt.figure(figsize=(8, 4))
sns.histplot(test_dff["Cosine_Similarity"], kde=True, color='purple')
plt.title("Cosine Similarity Distribution")
plt.xlabel("Cosine Similarity")
plt.ylabel("Frequency")
plt.show()

mean_cos = test_dff["Cosine_Similarity"].mean()
print("Mean Cosine Similarity:", mean_cos)
```

```python
import math
import torch
from tqdm.auto import tqdm

def calculate_ppl(texts):
    losses = []
    for text in tqdm(texts):
        inputs = tokenizer(text, return_tensors="pt").to(device)

        with torch.no_grad():
            outputs = model(**inputs, labels=inputs["input_ids"])

        loss = outputs.loss.item()
        losses.append(math.exp(loss))  # perplexity = e^loss

    return losses

test_dff["Perplexity"] = calculate_ppl(test_dff["Predicted_Response"].tolist())

mean_ppl = test_dff["Perplexity"].mean()
print("Mean Perplexity:", mean_ppl)
```
